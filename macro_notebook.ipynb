## MACROECONOMIC VARIABLES NOTEBOOK - ARGENTINA

The purpose of this notebook is to test the API that the government developed in order to facilitate access to public records related to macroeconomic indicators

```python
import api_min_hac as mh
import matplotlib.pyplot as plt
```

### Argentinean peso real exchange - Time series

```python
tcrm = mh.get_data(['116.3_TCRMA_0_M_36'], star_date='2002-01-01', limit=1000)
plt.plot(tcrm)
plt.title('REAL EXCHANGE RATE USD/ARS')
plt.show()
```

```python
tc_bna = mh.get_data(['168.1_T_CAMBIOR_D_0_0_26'], start_date='2008-01-01', limit= 1000)
tc_bna.plot()
plt.title('Official Exchange Rate Evolution 2015/2017')
plt.show()
```

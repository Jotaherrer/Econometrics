## Macro economic variables - Argentina

The idea of this notebook is to test the API that the government developed. I'm mainly reviewing macroeconomic and financial indicators

```python
import api_min_hac as mh
import matplotlib.pyplot as plt
```

### Argentinean peso adjusted by inflation time series (2002-2020)

```python
tcrm = mh.get_data(['116.3_TCRMA_0_M_36'], star_date='2002-01-01', limit=1000)
plt.plot(tcrm)
plt.title('REAL EXCHANGE RATE - USD/ARS')
plt.show()
```

### Argentinean peso - Nominal exchange rate time series (2015 - 2017)

```python
tc_bna = mh.get_data(['168.1_T_CAMBIOR_D_0_0_26'], start_date='2008-01-01', limit= 1000)
tc_bna.plot()
plt.title('NOMINAL EXCHANGE RATE - USD/ARS - 2015/2017')
plt.show()
```
